```python
from PIL import Image

# Загружаем изображение
image = Image.open("neurocreoru.png")
```

```python
# Проверяем, является ли изображение PNG
if image.format == "PNG":
    # Преобразуем изображение в RGBA (чтобы сохранить альфа-канал)
    image = image.convert('RGBA')
    
    # Создаем новую белую маску для непрозрачных пикселей
    white_mask = Image.new('L', image.size, 255)
    
    # Применяем маску к изображению
    image.putalpha(white_mask)
```

```python
# Сохраняем измененное изображение
image.save("neurocreoru_modified.png", format="PNG")
```
